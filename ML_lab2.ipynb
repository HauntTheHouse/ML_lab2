```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read train data
train_data = pd.read_csv("application_train.csv")
```

```python
# show data samples
max_col, max_row = train_data.shape
pd.set_option("display.max_rows", max_row)
pd.set_option("display.max_columns", max_col)
print(train_data.shape)
train_data.sample(7)
```

```python
# immediately delete column with index
train_data.drop("SK_ID_CURR", axis = 1, inplace = True)
```

```python
# reading data types
train_data.dtypes
```

```python
# we change data type to 'category' if type is 'object'
# or if it's starts with "FLAG" (accepts value '0' or '1')
# and another features that like this explicitly
for feature in train_data.columns:
    if train_data[feature].dtype == "object":
        train_data[feature] = train_data[feature].astype("category")
    elif feature.startswith("FLAG"):
        train_data[feature] = train_data[feature].astype("category")
    elif feature in {"REGION_RATING_CLIENT" ,"REG_REGION_NOT_LIVE_REGION" , "REGION_RATING_CLIENT_W_CITY", "REG_REGION_NOT_WORK_REGION", "LIVE_REGION_NOT_WORK_REGION", "REG_CITY_NOT_LIVE_CITY", "REG_CITY_NOT_WORK_CITY", "LIVE_CITY_NOT_WORK_CITY"}:
        train_data[feature] = train_data[feature].astype("category")
```

```python
# check empty cells
train_data.isnull().sum().sort_values(ascending = False)
```

```python
# plot data to decide how fill empty cells
nulls = train_data.select_dtypes(exclude='category').isnull().sum()
train_data[nulls[nulls > 0].index].hist(figsize=(20,20),color = "g", bins=30, xlabelsize=8, ylabelsize=8)
plt.show()
```

```python
percentage_of_empty_cells_in_features = train_data.isnull().sum()/train_data.shape[0]*100

# delete rows that have more than 50 empty cells
train_data = train_data.loc[train_data.isnull().sum(axis=1) < 50]

# delete cols that have more than 20% empty cells
cols_to_remove = set()
for feature in train_data.columns:
    if percentage_of_empty_cells_in_features[feature] > 20:
        cols_to_remove.add(feature)
        
train_data.drop(columns = cols_to_remove, axis=1, inplace=True)

# specially for this feature change n/a values to Unaccompanied
train_data["NAME_TYPE_SUITE"] = train_data["NAME_TYPE_SUITE"].fillna("Unaccompanied")

# firstly we want to process cols that have empty values
feature_with_low_empty_cells = set()
for feature in train_data.columns:
    if percentage_of_empty_cells_in_features[feature] > 0.0:
        feature_with_low_empty_cells.add(feature)
print(feature_with_low_empty_cells)
# if received features is categorical then rename cells to Unknown
# if feature is EXT_SOURCE_1 or EXT_SOURCE_2 then put median due to normal distribution
# if numeric - find arithmetic mean
for feature in feature_with_low_empty_cells:
    train_data[feature] = train_data[feature].replace(0, np.NaN)
    if train_data[feature].dtype.name == "category":
        train_data[feature] = train_data[feature].cat.add_categories("Unknown")
    elif feature == "EXT_SOURCE_1" or feature == "EXT_SOURCE_3":
        train_data[feature].fillna(train_data[feature].median(), inplace = True)
    else:
        train_data[feature].fillna(train_data[feature].mean(), inplace = True)
    
# print(test_data.shape)
```

```python
# look duplicates
train_data[train_data.duplicated()]
```

```python
# remove columns in which the same values are greater than 90 percent 
max_col, max_row = train_data.shape
for feature in train_data.columns:
    if train_data[feature].dtype.name == "category":
        count = train_data[feature].value_counts()
        if count[0]/max_col >= 0.90:
            print(count[0]/max_col)
            train_data.drop(feature, axis = 1, inplace = True)
```

```python
# read test data and process them like train data (but without deleting rows)
test_data = pd.read_csv("application_test.csv")

# also keep this column
TEST_ID = test_data["SK_ID_CURR"]
test_data.drop("SK_ID_CURR", axis = 1, inplace = True)

for feature in test_data.columns:
    if test_data[feature].dtype == "object":
        test_data[feature] = test_data[feature].astype("category")
    elif feature.startswith("FLAG"):
        test_data[feature] = test_data[feature].astype("category")
    elif feature in {"TARGET", "REGION_RATING_CLIENT" ,"REG_REGION_NOT_LIVE_REGION" , "REGION_RATING_CLIENT_W_CITY", "REG_REGION_NOT_WORK_REGION", "LIVE_REGION_NOT_WORK_REGION", "REG_CITY_NOT_LIVE_CITY", "REG_CITY_NOT_WORK_CITY", "LIVE_CITY_NOT_WORK_CITY"}:
        test_data[feature] = test_data[feature].astype("category")

# specially for this feature change n/a values to Unaccompanied
test_data["NAME_TYPE_SUITE"] = test_data["NAME_TYPE_SUITE"].fillna("Unaccompanied")

# firstly we want to process cols that have empty values
feature_with_low_empty_cells = set()
for feature in test_data.columns:
    if percentage_of_empty_cells_in_features[feature] > 0.0:
        feature_with_low_empty_cells.add(feature)

# if received features is categorical then rename cells to Unknown
# if feature is EXT_SOURCE then put median due to normal distribution
# if numeric - find arithmetic mean
for feature in feature_with_low_empty_cells:
    test_data[feature] = test_data[feature].replace(0, np.NaN)
    if test_data[feature].dtype.name == "category":
        test_data[feature] = test_data[feature].cat.add_categories("Unknown")
    elif feature == "EXT_SOURCE_1" or feature == "EXT_SOURCE_3":
        test_data[feature].fillna(test_data[feature].median(), inplace = True)
    else:
        test_data[feature].fillna(test_data[feature].mean(), inplace = True)
        
test_data = pd.get_dummies(test_data, drop_first=True)

# now we have 223 features and we want to keep only 21 of them
features_to_keep = train_data.columns
# print(len(features_to_keep))

max_col, max_row = test_data.shape
print(test_data.shape)

for feature in test_data.columns:
    if feature not in features_to_keep:
        test_data.drop(feature, axis = 1, inplace = True)
            
print(test_data.shape)
test_data.sample(7)
```

```python
# import libraries for building the classifier 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, accuracy_score
```

```python
# split train table into train and validation data
train_x, val_x, train_y, val_y = train_test_split(train_data, TARGET_COL, test_size=0.2)
print(f"Train: {train_x.shape}, {train_y.shape}")
print(f"Validation: {val_x.shape}, {val_y.shape}")
```

```python
# normalize columns
sc = StandardScaler()
scaled_train_x = sc.fit_transform(train_x)
scaled_val_x = sc.transform(val_x)
```

```python
# see how logistic regression works
lr = LogisticRegression(class_weight="balanced")
lr.fit(scaled_train_x, train_y)
train_preds = lr.predict(scaled_train_x)
print(f"Train ROC_AUC: {roc_auc_score(train_y, train_preds)}")
val_preds = lr.predict(scaled_val_x)
print(f"Validation ROC_AUC: {roc_auc_score(val_y, val_preds)}")
```

```python
# see how decition tree classifier works
dtc = DecisionTreeClassifier(class_weight="balanced", max_depth=10)
dtc.fit(scaled_train_x, train_y)
train_preds = dtc.predict(scaled_train_x)
print(f"Train ROC_AUC: {roc_auc_score(train_y, train_preds)}")
val_preds = dtc.predict(scaled_val_x)
print(f"Validation ROC_AUC: {roc_auc_score(val_y, val_preds)}")
```

```python
# normalize data from the test set 
scaled_test_x = sc.transform(test_data)
```

```python
# get probabilities of predictions 
test_preds = dtc.predict_proba(scaled_test_x)
```

```python
# put our result to submission.csv
pd.DataFrame({
    'SK_ID_CURR': TEST_ID,
    'TARGET':     [ row[1] for row in test_preds]
}).to_csv('./submission.csv',index=False)
```
